```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
# selecting the columns I want to work with, and assigning them to a new df movies.

movies = df[['title','genres','original_language','release_date','budget','revenue', 'runtime','popularity','vote_average','vote_count']]
```

```python
#  The genres column has a lot of extra info. I just want the category the movie falls into so I cut the extra info out. Saving
#  saving the new info as genre and then removing the old column. 

gen = movies['genres'].apply(lambda genres: genres.split('"')[5:6])
```

```python
movies['genre'] = gen
```

```python
movies.drop(['genres'],axis=1, inplace=True)
```

```python
movies['genre'] = movies['genre'].apply(lambda g: str(g))
```

## About this data set
This data set was scraped from the webpage Imdb.  It is the break down of just under 5000 movies. I have removed a handful of columns to focus on specific data. I have also change the genres col to genre while removing the ids and just taking the categories name.

## Hypotheses

1) What columns have the most influence on revenue and vote_average?

2) How does vote_count corrilate to popularity?

3) Does runtime increase budget and do certain genres tend to run longer than others?

### 1) What columns have the most influence on revenue and vote_average?

From our heatmap, I am going to take the 3 'hotter' tiles from Revenue.  These three tiles should have a more dominant trend.

```python
plt.figure(figsize=(16, 12))

# row one over view of positive corrilations 
plt.subplot(2,3,1)
plt.scatter(x='revenue', y='budget',c='blue',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Budget in Hundres of Million $')
plt.title('Budget to Revenue Scatter Chart')

plt.subplot(2,3,2)
plt.scatter(x='revenue', y='popularity',c='green',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Popularity')
plt.title('Popularity Vs Revenue Scatter Chart')

plt.subplot(2,3,3)
plt.scatter(x='revenue', y='vote_count', c='red',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Votes')
plt.title('Number of Votes Vs Revenue Scatter Chart')

# row two close up of previous charts
plt.subplot(2,3,4)
plt.scatter(x='revenue', y='budget',s= 10, c='blue',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Budget in Hundres of Million $')
plt.xlim(-10000,1600000000)
plt.ylim(0,300000000)
plt.title('Budget to Revenue Close Up')
print(ttest_ind(movies['budget'],movies['revenue'], equal_var=False))

plt.subplot(2,3,5)
plt.scatter(x='revenue', y='popularity',s= 10, c='green',data=movies)
plt.xlim(-10000,1600000000)
plt.ylim(0,300)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Popularity')
plt.title('Popularity Vs Revenue Close Up')

plt.subplot(2,3,6)
plt.scatter(x='revenue', y='vote_count',s= 10, c='red',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Votes')
plt.title('Number of Votes Vs Revenue Close up')
plt.xlim(-10000,1600000000)
plt.ylim(0,12000)


plt.tight_layout
plt.show()
```

```python
print((ttest_ind(movies['budget'],movies['revenue'], equal_var=False)),
    (ttest_ind(movies['popularity'],movies['revenue'], equal_var=False)),
    (ttest_ind(movies['vote_count'],movies['revenue'], equal_var=False)))
```

 Now lets go head and plot the two remaining tiles to see how they compare to our first three.
 We have runtime with 0.25 and average vote with 0.2.

```python
plt.figure(figsize=(20,20))

plt.subplot(2,2,1)
plt.scatter(x='revenue', y='runtime',c='orange',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Lenght in Mins')
plt.title('Score Vs Revenue Scatter Plot')

plt.subplot(2,2,2)
plt.scatter(x='revenue', y='vote_average',c='purple',data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Average Score')
plt.title('Length Vs Revenue Scatter Plot')

plt.subplot(2,2,3)
plt.scatter(x='revenue', y='runtime',c='orange',s=10,data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Lenght in Mins')
plt.title('Score Vs Revenue Scatter Plot')
plt.xlim(-10000,1600000000)
plt.ylim(45,250)

plt.subplot(2,2,4)
plt.scatter(x='revenue', y='vote_average',c='purple',s=10,data=movies)
plt.xlabel('Revenue in Billions $')
plt.ylabel('Average Score')
plt.title('Length Vs Revenue Scatter Plot')
plt.xlim(-10000,1600000000)
plt.ylim(2,10)

plt.tight_layout
plt.show()
```

```python
print((ttest_ind(movies['runtime'],movies['revenue'], equal_var=False)))
print((ttest_ind(movies['vote_average'],movies['revenue'], equal_var=False)))
```

As we compare these groups of charts, we can see that the first 3 have a much stronger tendency to move up along the Y-Axis as revenue is increased.  The last two charts are influence a lot less by an increase in revenue.  We can also see this by looking at the distribution of runtime and average_score.
